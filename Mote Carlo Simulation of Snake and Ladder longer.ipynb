## Implement Monte Carlo Randomized Algorithm Using Snakes and ladder game


### Problem Statement
The whole program is a simulation of the popular game **snakes and ladder**. We are trying to find out the average shortest number of moves it took to complete the whole game using a _randomized algorithm_ , **Monte carlo**. 
While finding the minimum number of moves of the simulation made we also trying to find out the **shortest path, longest length and average length.**
We observed outputs for all these questions by **simulating this game for multiple times.**

```python
#Import Libraries
import random #random Library
import math #math Library
import matplotlib.pyplot as plt #Imports Pyplot class from matplotlib as plt 
import numpy as np
import scipy.stats as stats
```

## Snake and ladder board we used for the simulation


```python
#We used Dictionaries to represent the values and transitionin the board when ever it hits snake or ladder.

SnaKEsLadder = {53:2,8:77,37:94,58:82,99:12,87:57,90:43,72:30}
```

```python

trials = [] # the array where the length of each game is appended  
GameN = 0 # initializing the simulation
NGames = 10000 # the No. of simulations are made in this module
position = [] #Each position it reached in a game

# the loop initialized here for each simulation,

while GameN< NGames: 
	GameN = GameN+1 #counter for each simulation of game, runs untill all NGames are simulated
	Myposition=0 # initial position of the computer 
	num = 0 #counter initiation for number of moves
	while Myposition < 100: #runs till the position reaches to 30
		num = num+1 #counter initiation for number of moves appended in trials
		Myposition = Myposition+random.randint(1,6) #by using random function to Roll dice and assigning it to the present state
		if( Myposition in SnaKEsLadder): #when met a ladder or snake
			Myposition = SnaKEsLadder[Myposition]
		if(Myposition >100):
    			Myposition = 100	
		position.append(Myposition) #each move made by the computer is registered in position  
	trials.append(num) 
```

```python
# print a list of game lengths
print(trials)
```

```python

#Print average No of length
AllGamesTotal = 0
for i in trials:
	AllGamesTotal = AllGamesTotal+i
print(AllGamesTotal/NGames)
```

```python
# maximum moves and Minimum Moves made
LonGame = max(trials)
ShotGame = min(trials)

print("Maximum length:", LonGame,"Minimum length:",ShotGame)
```

```python
#Find the shortest and longest game index
SPlace = trials.index(ShotGame)
LPlace = trials.index(LonGame)
#print(SPlace,LPlace)
```

```python
# Using the game length we organize a matrix for each move the computer visited in each game
spli = []
def splimove(lenn,val):
    count=0
    for i in lenn:
        lis =[]
        for j in range(i):
            lis.append(val[count+j])
        count+=i
        spli.append(lis)
  
   
splimove(trials,position)# we now have positions made in each game 

print(spli)
```

```python
#Finding the shortest path and Longest path
Spath = spli[SPlace]
Lpath = spli[LPlace]
print("Shortest path :",Spath,"longest Path :",Lpath)
```

```python
# A line graph for trials
plt.xlim(1,len(trials)-1) 
plt.plot(trials)
plt.plot()
plt.show()
```

```python
# A boxPlot of Length of each game
SortLen = sorted(trials)
plt.boxplot(SortLen,vert=0,patch_artist=True)
plt.show()
```

```python
# A normal distribution curve to see the most probable output in the simulation made.
fit = stats.norm.pdf(SortLen, np.mean(SortLen), np.std(SortLen))

plt.plot(SortLen,fit,'-o')

plt.show()              
```

```python
# A sorted line graph to see the mode the most repeated values.
plt.xlim(1,len(SortLen)-1) 
plt.plot(SortLen)
plt.plot()
plt.show()
```
